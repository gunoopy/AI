# Anomaly Detection using Autoencoder
> by Conv Layer

<br/>
<br/>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf

tf.__version__
```

<br/>
<br/>
<br/>
<br/>

## Data

```python
master_url_root = "https://raw.githubusercontent.com/numenta/NAB/master/data/"

df_small_noise_url_suffix = "artificialNoAnomaly/art_daily_small_noise.csv"
df_small_noise_url = master_url_root + df_small_noise_url_suffix
df_small_noise = pd.read_csv(
    df_small_noise_url, parse_dates=True, index_col="timestamp"
)

df_daily_jumpsup_url_suffix = "artificialWithAnomaly/art_daily_jumpsup.csv"
df_daily_jumpsup_url = master_url_root + df_daily_jumpsup_url_suffix
df_daily_jumpsup = pd.read_csv(
    df_daily_jumpsup_url, parse_dates=True, index_col="timestamp"
)
```

```python
display(df_small_noise.head(), df_daily_jumpsup.head())
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, sharex = True, sharey = True, figsize = (12,8))

ax1.plot(df_small_noise, c = 'gray')
ax2.plot(df_daily_jumpsup, c = 'gray')
plt.show()
```

<br/>
<br/>
<br/>
<br/>

## Scaling

```python
X_train_mean = df_small_noise.mean()
X_train_std = df_small_noise.std()
X_train = (df_small_noise - X_train_mean) / X_train_std

X_train.shape
```

<br/>
<br/>
<br/>
<br/>

## Divide by step

```python
TIME_STEPS = 288

def create_seq(values, time_steps = TIME_STEPS) :
    output = []
    for i in range(len(values) - time_steps + 1) :
        output.append(values[i : (i + time_steps)])
    return np.stack(output)

X_train = create_seq(X_train)
X_train.shape
```

<br/>
<br/>
<br/>
<br/>

## Modeling

```python
from tensorflow.keras import layers, models
from tensorflow.keras.optimizers import Adam

model = models.Sequential()

#
# Encoder
#

model.add(layers.Conv1D(filters = 32, kernel_size = 7, padding = 'same', strides = 2, activation = 'relu',
                        input_shape = (X_train.shape[1], X_train.shape[2])))

model.add(layers.Dropout(0.2))

model.add(layers.Conv1D(filters = 16, kernel_size = 7, padding = 'same', strides = 2, activation = 'relu'))


#
# Decoder
#

model.add(layers.Conv1DTranspose(filters = 16, kernel_size = 7, padding = 'same', strides = 2, activation = 'relu'))

model.add(layers.Dropout(0.2))

model.add(layers.Conv1DTranspose(filters = 32, kernel_size = 7, padding = 'same', strides = 2, activation = 'relu'))

model.add(layers.Conv1DTranspose(filters = 1, kernel_size = 7, padding = 'same'))


model.compile(optimizer = Adam(learning_rate = 0.001),
              loss = 'mse')

model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping

es = EarlyStopping(monitor = 'val_loss',
                   patience = 5,
                   mode = 'min',
                   verbose = 1)

hist = model.fit(X_train, X_train,
                epochs = 50,
                batch_size = 128,
                validation_split = 0.1,
                callbacks = [es])
```

```python
plt.figure(figsize = (8,6))
plt.plot(hist.history['loss'], label = 'train')
plt.plot(hist.history['val_loss'], label = 'valid')
plt.legend()
plt.title('Loss')
plt.show()
```

<br/>
<br/>
<br/>
<br/>

## MAE & Threshold

```python
pred = model.predict(X_train)
mae = np.mean(np.abs(pred - X_train), axis = 1)

plt.figure(figsize = (8,6))

plt.hist(mae, bins = 100)
plt.xlabel('mae')
plt.ylabel('No of samples')
plt.title('MAE of train data')
plt.show()

threshold = np.max(mae)
print(f'Reconstruction error threshold : {threshold}')
```

<br/>
<br/>
<br/>
<br/>

## Prediction

```python
np.random.seed(42)
index = np.random.randint(0, X_train.shape[0], 3)

fig, axes = plt.subplots(3, 1, sharex = True, sharey = True, figsize = (12, 10))

fig.suptitle('Prediction', y = 0.93, fontsize = 20)

for i, idx in enumerate(index) :
    axes[i].plot(X_train[idx], color = 'gray', label = 'data')
    axes[i].plot(pred[idx], color = 'red', label = 'prediction')
    axes[i].legend()

plt.show()
```

<br/>
<br/>
<br/>
<br/>

## Scaling & Divide by step

```python
# Scaling
X_test = (df_daily_jumpsup - X_train_mean) / X_train_std
plt.figure(figsize = (12, 4))
plt.plot(X_test, color = 'gray')
plt.show()

# Divide by step
X_test = create_seq(X_test.values)
X_test.shape
```

<br/>
<br/>
<br/>
<br/>

## Prediction & MAE

```python
test_pred = model.predict(X_test)
test_mae = np.mean(np.abs(test_pred - X_test), axis = 1)
test_mae = test_mae.reshape(-1)

plt.figure(figsize = (8,6))
plt.hist(test_mae, bins = 100)
plt.xlabel('mae')
plt.ylabel('No of samples')
plt.title('MAE of test data')
plt.show()
```

<br/>
<br/>
<br/>
<br/>

## Get Anomalies

```python
anomalies = test_mae > threshold
print(f'Number of anomaly samples : {np.sum(anomalies)}')
print(f'Indices of anomaly samples : {np.where(anomalies)}')
```

```python
anomalous_data_indices = set()
for idx in np.where(anomalies)[0] :
    anomalous_data_indices = anomalous_data_indices | set(range(idx, idx + TIME_STEPS))
    
anomalous_data_indices = list(anomalous_data_indices)

len(anomalous_data_indices)
```

```python
anomalous_data_indices = []

for data_idx in range(TIME_STEPS - 1, len(df_daily_jumpsup) - TIME_STEPS + 1):
    if np.all(anomalies[data_idx - TIME_STEPS + 1 : data_idx]):
        anomalous_data_indices.append(data_idx)
        
df_subset = df_daily_jumpsup.iloc[anomalous_data_indices]

plt.figure(figsize = (12, 4))
plt.plot(df_daily_jumpsup, color = 'gray', label = 'data')
plt.plot(df_subset, color = 'red', label = 'anomaly')
plt.legend()
plt.title('Anonmaly Data', fontsize = 20, y = 1.1)
plt.show()
```
